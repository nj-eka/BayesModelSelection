```python
import numpy as np

import scipy.special as sp

import matplotlib.pyplot as plt
```

```python
beta = sp.gamma(0.5)*sp.gamma(0.5)/sp.gamma(1)
```

```python
b = np.random.rand()*10
```

```python
np.random.seed(1224)
```

```python
N = 365
x = np.linspace(1, N, N)

y = np.zeros(N)+ b
```

```python
for i in range(1, N):
    y[i] = np.random.normal(loc = y[i-1], scale=1)
```

```python
plt.plot(x,y)

plt.grid()
plt.ylabel('x')
plt.xlabel('t')

# plt.savefig('first.pdf')
plt.show()
```

```python
x = np.linspace(0, 1, 100)

y = 6*x*(1-x)
```

```python
plt.plot(x,y, label = 'p(x)=Beta(2,2)')

plt.grid()
plt.legend(loc = 'best')
plt.ylabel('p(x)')
plt.xlabel('x')

# plt.savefig('second.pdf')
plt.show()
```

```python
x = np.linspace(-100, 100, 1000)

y1 = np.ones(shape = [x.shape[0]])

y2 = np.zeros(shape = [x.shape[0]])
y2[x >= 0] += 2
```

```python
plt.plot(x, y1, label = 'p(mu)')
plt.plot(x, y2, label = 'p(sigma)')

plt.grid()
plt.legend(loc = 'best')
plt.ylabel('p(x)')
plt.xlabel('x')

plt.savefig('third.pdf')
plt.show()
```

```python
x = np.linspace(-1, 2, 1000)

y = np.ones(shape = [x.shape[0]])
y[x < 0] = 0
y[x>1] = 0
```

```python
plt.plot(x, y, label = 'p(theta)')

plt.grid()
plt.legend(loc = 'best')
plt.ylabel('p(x)')
plt.xlabel('x')

plt.savefig('forth.pdf')
plt.show()
```
